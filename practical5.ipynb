```
#Smoothing
#Linear operation
import cv2
import numpy as np
from matplotlib import pyplot as plt
img = cv2.imread('/content/drive/My Drive/cv/10.jpeg')
kernel = np.ones((5,5),np.float32)/25
dst = cv2.filter2D(img,-1,kernel)
plt.subplot(121),plt.imshow(img),plt.title('Original')
plt.xticks([]),plt.yticks([])
plt.subplot(122),plt.imshow(dst),plt.title('Linear Operation')
plt.xticks([]),plt.yticks([])
plt.show()
```

```
#Median operation
import cv2
import numpy as np
from matplotlib import pyplot as plt
img = cv2. imread('/content/drive/My Drive/cv/10.jpeg')
median = cv2.medianBlur(img,5)
plt.subplot(121),plt.imshow(img),plt.title('Original')
plt. xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(median),plt.title('Median')
plt.xticks([]), plt.yticks ([])
plt.show()
```

```
#Averaging Operation
import cv2
import numpy as np
from matplotlib import pyplot as plt
img1 = cv2.imread('/content/drive/My Drive/cv/10.jpeg')
blur = cv2.blur(img1,(5,5))
plt.subplot(121),plt.imshow(img1),plt.title('Original')
plt. xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(blur),plt.title('Averaging')
plt.xticks([]), plt.yticks ([])
plt.show()
```
