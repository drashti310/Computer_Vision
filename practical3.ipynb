```
#AIM:-Convert Image from one colourspace to another as mentioned below
#BGR ↔ Gray 

import cv2
import numpy as np
from matplotlib import pyplot as plt

orignal_img = cv2.imread('/content/drive/My Drive/cv/images.jpg')
img = cv2.cvtColor(orignal_img, cv2.COLOR_BGR2GRAY)
plt.subplot(1,1,1),plt.imshow(img)
plt.title('GrayScale')
```

```


img = cv2.cvtColor(orignal_img,cv2.COLOR_BGR2YCrCb)
plt.subplot(1,1,1),plt.imshow(img)
plt.title('YCrCb')
```

```
#BGR ↔ HSV
img = cv2.cvtColor(orignal_img,cv2.COLOR_BGR2HSV)
plt.subplot(1,1,1),plt.imshow(img)
plt.title('HSV color space')
```

```
#Python program to read image as LAB color space
img = cv2.cvtColor(orignal_img,cv2.COLOR_BGR2LAB)
plt.subplot(1,1,1),plt.imshow(img)
plt.title('LAB color space')
```

```
#Python program to read image as Logtransformation
img_log = (np.log(img+1)/(np.log(1+np.max(orignal_img))))*255
# Specify the data type
img_log = np.array(img_log,dtype=np.uint8)
# Display the image
plt.subplot(1,1,1),plt.imshow(img_log)
plt.title('Logtransformation')
```
